## 행 추가

```python
df = pd.DataFrame({
    "이름": ["Evan", "Sara", "Chris", "Messi", "Woods", "Jordan"], 
    "국어": [70, 90, None, None, 100, 85], 
    "영어": [65, 100, None, 90, 70, 80]
})

df
```

```python
df.loc[6, "이름":"영어"] = ["Kim", 50, 60]
df
```

```python
# 국어 점수가 90점 이상이면 "A"
# 국어 점수가 80~89 "B"
# 국어점수가 70~79 "C"
# 나머지는 F
# df['등급'] =   사용하기
```

1. 데이터가 주어졌고, pandas 데이터 프레임으로 변경되어 df로 지정된 상태로 주어짐
2. 우리가 만드려는 것은 국어 점수에 따른 등급을 만들어보려고함
3. 국어 점수에 따른 등급을 나타내는 조건문을 지정해야한다.
4. 국어점수로 나타낸 등급을 데이터 프레임에 적용시킨다.

```python
def grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F"

df['등급'] = df['국어'].apply(grade)
df
```

## 다른 방법

```python
condition_list = [(df["국어"] >= 90), (df['국어'] >= 80) & (df['국어'] < 90), (df['국어'] >= 70) & (df['국어'] < 80)]
choice_list = ["A", "B", "C"]
df['점수'] = np.select(condition_list, choice_list, default = "F")
df
```

- numpy.where, 조건식 한번만 적용

if 조건식, 참, 거짓
