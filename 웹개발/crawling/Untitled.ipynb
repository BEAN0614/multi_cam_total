```python
import requests
import json
import pandas as pd

SERVICE_KEY = '564b7852646a686a34336f4f6c5571'
base_url = 'http://openapi.seoul.go.kr:8088/'

# 각 데이터프레임의 크기
chunk_size = 1000

# 전체 데이터를 가져올 범위 설정
start_index = 1
end_index = 12000

# 데이터를 저장할 리스트 초기화
dataframes = []

# 시작 인덱스부터 끝 인덱스까지 데이터를 가져와서 리스트에 추가
for i in range(start_index, end_index + 1, chunk_size):
    # 요청할 URL 생성
    url = f'{base_url}{SERVICE_KEY}/json/tbLnOpendataRtmsV/{i}/{i + chunk_size - 1}/2023/11560'
    req = requests.get(url)
    
    # JSON 데이터 추출
    content = req.json()
    
    # 결과가 있는 경우 데이터프레임 생성 및 리스트에 추가
    if 'tbLnOpendataRtmsV' in content:
        result = pd.DataFrame(content['tbLnOpendataRtmsV']['row'])
        dataframes.append(result)

# 모든 데이터프레임을 하나로 합치기
final_df = pd.concat(dataframes, ignore_index=True)

# 결과 확인
print(final_df)
```

```python
import pandas as pd
import requests

# 특정 인증서 파일을 사용하여 SSL 인증서 검증
response = requests.get(url, verify='/path/to/ca-cert.pem')

url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson?serviceKey=FPSnL7hBknn%2B4ogFEqKq20afCM0Q6EcTRjCefdwtGKJa3qF8IG8NJMab9H6RZvUCyynUp6%2BjdEzpImofdHX9zA%3D%3D&pageNo=1&numOfRows=1000&startCreateDt=20210101'



# API 인증키
SERVICE_KEY = 'FPSnL7hBknn%2B4ogFEqKq20afCM0Q6EcTRjCefdwtGKJa3qF8IG8NJMab9H6RZvUCyynUp6%2BjdEzpImofdHX9zA%3D%3D'

# 요청할 URL 생성 함수
def generate_url(page_no):
    base_url = 'https://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson'
    return f'{base_url}?serviceKey={SERVICE_KEY}&pageNo={page_no}&numOfRows=1000&startCreateDt=20210101'

# 데이터 수집 및 저장
all_data = []
page_no = 1

while True:
    url = generate_url(page_no)
    response = requests.get(url)
    data = response.json()
    items = data['response']['body']['items']
    
    if not items:  # 데이터가 없으면 종료
        break
    
    all_data.extend(items)
    print(f'Page {page_no}: Fetched {len(items)} items')
    
    page_no += 1

# 데이터프레임으로 변환하여 CSV 파일로 저장
df = pd.DataFrame(all_data)
df.to_csv('result.csv', index=False)

print('Data collected and saved to result.csv')
```
